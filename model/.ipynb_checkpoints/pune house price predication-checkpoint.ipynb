```python
# we have value like in term of ranges  by taking mean of that range we can convert into value
# we are using the code
total_squre_fit=[]
for str_value in df["total_sqft"]:
    try:
        total_squre_fit.append(float(str_value))
    except:
        try:
            temp=[]
            temp=str_value.split("-")
            total_squre_fit.append((float(temp[0])+float(temp[-1]))/2)
        except:
            total_squre_fit.append(np.nan)
```

### train and test split

```python
# defnning the independent and dependent variable
x=df.drop("price",axis=1)
y=df["price"]
```

```python
## train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=25)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
# excluding the object 
df=df.select_dtypes(exclude=object)
```

#### model training

```python
# creating the model instance
model=LinearRegression()
```

```python
# model fitting
model.fit(x_train,y_train)
```

```python
### evalution of model on testing data
y_pred = model.predict(x_test)

mse = mean_squared_error(y_pred,y_test)
print("MSE :",mse)

rmse = np.sqrt(mse)
print("RMSE :",rmse)

mae = mean_absolute_error(y_pred,y_test)
print("MAE :",mae)

r2 = r2_score(y_pred,y_test)
print("R Squared :",r2)
```

```python
### evalution of model on training data
y_pred_train = model.predict(x_train)

mse = mean_squared_error(y_pred_train, y_train)
print("MSE :",mse)

rmse = np.sqrt(mse)
print("RMSE :",rmse)

mae = mean_absolute_error(y_pred_train, y_train)
print("MAE :",mae)

r2 = r2_score(y_pred_train, y_train)
print("R Squared :",r2)
```
